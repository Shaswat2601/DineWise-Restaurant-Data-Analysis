```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('dataSci.csv')

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Percentage of Restaurants Offering Table Booking and Online Delivery
# Total number of restaurants
total_restaurants = len(data)

# Count restaurants offering table booking and online delivery
table_booking_count = data['Has Table booking'].value_counts().get('Yes', 0)
online_delivery_count = data['Has Online delivery'].value_counts().get('Yes', 0)

# Calculate percentages
table_booking_percentage = (table_booking_count / total_restaurants) * 100
online_delivery_percentage = (online_delivery_count / total_restaurants) * 100

print(f'Percentage of Restaurants Offering Table Booking: {table_booking_percentage:.2f}%')
print(f'Percentage of Restaurants Offering Online Delivery: {online_delivery_percentage:.2f}%')

# Visualization of Table Booking and Online Delivery
plt.figure(figsize=(8, 5))
sns.barplot(x=['Table Booking', 'Online Delivery'], y=[table_booking_percentage, online_delivery_percentage], palette='viridis')
plt.title('Percentage of Restaurants Offering Table Booking and Online Delivery')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)
plt.show()

# 2. Average Ratings Comparison
# Calculate average ratings for restaurants with and without table booking
average_rating_with_booking = data[data['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_booking = data[data['Has Table booking'] == 'No']['Aggregate rating'].mean()

print(f'Average Rating of Restaurants with Table Booking: {average_rating_with_booking:.2f}')
print(f'Average Rating of Restaurants without Table Booking: {average_rating_without_booking:.2f}')

# Visualization of Average Ratings
plt.figure(figsize=(8, 5))
sns.barplot(x=['With Table Booking', 'Without Table Booking'], y=[average_rating_with_booking, average_rating_without_booking], palette='mako')
plt.title('Average Ratings Comparison')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.show()

# 3. Online Delivery Availability Among Different Price Ranges
# Create a new column for price range categories
def categorize_price_range(price):
    if price < 500:
        return 'Low'
    elif 500 <= price < 1500:
        return 'Medium'
    else:
        return 'High'

data['Price Range Category'] = data['Average Cost for two'].apply(categorize_price_range)

# Calculate the percentage of online delivery by price range
delivery_by_price_range = data.groupby('Price Range Category')['Has Online delivery'].value_counts(normalize=True).unstack() * 100

# Visualization of Online Delivery by Price Range
delivery_by_price_range.plot(kind='bar', stacked=True, color=['lightblue', 'salmon'], figsize=(10, 6))
plt.title('Online Delivery Availability by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Online Delivery', labels=['No', 'Yes'])
plt.ylim(0, 100)
plt.show()

# Displaying the results in a more readable format
for price_range, row in delivery_by_price_range.iterrows():
    print(f"\nPrice Range: {price_range}")
    for delivery_option, percentage in row.items():
        print(f"  {delivery_option}: {percentage:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('dataSci.csv')

# Set the style for seaborn
sns.set(style="whitegrid")

# 1. Determine the Most Common Price Range
# Create a new column for price range categories
def categorize_price_range(price):
    if price < 500:
        return 'Low'
    elif 500 <= price < 1500:
        return 'Medium'
    else:
        return 'High'

data['Price Range Category'] = data['Average Cost for two'].apply(categorize_price_range)

# Most common price range
most_common_price_range = data['Price Range Category'].mode()[0]
print(f'Most Common Price Range: {most_common_price_range}')

# 2. Calculate the Average Rating for Each Price Range
average_rating_by_price_range = data.groupby('Price Range Category')['Aggregate rating'].mean()

print("\nAverage Rating for Each Price Range:")
print(average_rating_by_price_range)

# Visualization of Average Ratings by Price Range
plt.figure(figsize=(8, 5))
sns.barplot(x=average_rating_by_price_range.index, y=average_rating_by_price_range.values, palette='viridis')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.show()

# 3. Identify the Color Representing the Highest Average Rating
# Find the price range with the highest average rating
highest_average_rating = average_rating_by_price_range.max()
highest_price_range = average_rating_by_price_range.idxmax()

# Display the result
print(f'\nPrice Range with the Highest Average Rating: {highest_price_range} ({highest_average_rating:.2f})')

# Visualization of Average Ratings with Color Representation
plt.figure(figsize=(8, 5))
colors = ['lightblue' if x != highest_price_range else 'salmon' for x in average_rating_by_price_range.index]
sns.barplot(x=average_rating_by_price_range.index, y=average_rating_by_price_range.values, palette=colors)
plt.title('Average Rating by Price Range with Highlight')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.axhline(y=highest_average_rating, color='red', linestyle='--', label='Highest Average Rating')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('dataSci.csv')

# Preprocessing: Encode categorical variables
data['Has Table Booking'] = data['Has Table booking'].map({'Yes': 1, 'No': 0})
data['Has Online Delivery'] = data['Has Online delivery'].map({'Yes': 1, 'No': 0})

# Feature Extraction: Length of Restaurant Name and Address
data['Name Length'] = data['Restaurant Name'].apply(len)
data['Address Length'] = data['Address'].apply(len)

# Select features and target variable
features = data[['Average Cost for two', 'Has Table Booking', 'Has Online Delivery', 'Name Length', 'Address Length']]
target = data['Aggregate rating']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 1. Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)

# Evaluate Linear Regression
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
print(f'Linear Regression - MSE: {mse_linear:.2f}, R^2: {r2_linear:.2f}')

# 2. Decision Tree Regressor
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)

# Evaluate Decision Tree
mse_tree = mean_squared_error(y_test, y_pred_tree)
r2_tree = r2_score(y_test, y_pred_tree)
print(f'Decision Tree - MSE: {mse_tree:.2f}, R^2: {r2_tree:.2f}')

# 3. Random Forest Regressor
forest_model = RandomForestRegressor(random_state=42)
forest_model.fit(X_train, y_train)
y_pred_forest = forest_model.predict(X_test)

# Evaluate Random Forest
mse_forest = mean_squared_error(y_test, y_pred_forest)
r2_forest = r2_score(y_test, y_pred_forest)
print(f'Random Forest - MSE: {mse_forest:.2f}, R^2: {r2_forest:.2f}')

# 4. Comparison of Model Performance
models = ['Linear Regression', 'Decision Tree', 'Random Forest']
mse_values = [mse_linear, mse_tree, mse_forest]
r2_values = [r2_linear, r2_tree, r2_forest]

# Visualization of Model Performance
plt.figure(figsize=(12, 6))
sns.barplot(x=models, y=mse_values, palette='viridis')
plt.title('Model Performance Comparison (MSE)')
plt.ylabel('Mean Squared Error')
plt.ylim(0, max(mse_values) + 10)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=models, y=r2_values, palette='mako')
plt.title('Model Performance Comparison (R^2 Score)')
plt.ylabel('R^2 Score')
plt.ylim(0, 1)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('dataSci.csv')

# 1. Analyze the relationship between the type of cuisine and the restaurant's rating
# Assuming 'Cuisines' and 'Aggregate rating' are the relevant columns
# Explode the cuisines if there are multiple cuisines listed
data['Cuisines'] = data['Cuisines'].str.split(', ')
data_exploded = data.explode('Cuisines')

# Calculate average rating by cuisine
average_rating_by_cuisine = data_exploded.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

# Visualization of Average Ratings by Cuisine
plt.figure(figsize=(12, 6))
sns.barplot(x=average_rating_by_cuisine.index, y=average_rating_by_cuisine.values, palette='viridis')
plt.title('Average Ratings by Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.ylim(0, 5)
plt.show()

# 2. Identify the most popular cuisines based on the number of votes
# Assuming 'Votes' is the relevant column
popular_cuisines = data_exploded['Cuisines'].value_counts().head(10)

# Visualization of Most Popular Cuisines
plt.figure(figsize=(12, 6))
sns.barplot(x=popular_cuisines.index, y=popular_cuisines.values, palette='mako')
plt.title('Most Popular Cuisines Based on Number of Votes')
plt.xlabel('Cuisine')
plt.ylabel('Number of Votes')
plt.xticks(rotation=45)
plt.show()

# 3. Determine if there are specific cuisines that tend to receive higher ratings
# Calculate the average rating and number of votes for each cuisine
cuisine_rating_votes = data_exploded.groupby('Cuisines').agg(
    Average_Rating=('Aggregate rating', 'mean'),
    Total_Votes=('Votes', 'sum')
).reset_index()

# Visualization of Average Rating vs Total Votes
plt.figure(figsize=(12, 6))
sns.scatterplot(data=cuisine_rating_votes, x='Total_Votes', y='Average_Rating', hue='Cuisines', palette='tab10', legend=None)
plt.title('Average Rating vs Total Votes by Cuisine')
plt.xlabel('Total Votes')
plt.ylabel('Average Rating')
plt.axhline(y=cuisine_rating_votes['Average_Rating'].mean(), color='red', linestyle='--', label='Average Rating')
plt.legend()
plt.show()
```
